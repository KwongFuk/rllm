```python
import torch

# 加载 .pt 文件（确保信任来源）
data = torch.load('/home/yue/guangfu/trajectories/mcp_tavily/trajectories.pt', weights_only=False)

# 只看第一条轨迹
traj = data[0]

print("🧠 Question:", traj.task.get("question"))
print("✅ Ground Truth:", traj.task.get("ground_truth"))
print("📦 Data Source:", traj.task.get("data_source"))
print("📍 Number of Steps:", len(traj.steps))
print("=" * 50)

# 查看前两步内容
for i, step in enumerate(traj.steps[:2]):
    print(f"\n--- Step {i + 1} ---")
    
    if hasattr(step, "chat_completions"):
        for msg in step.chat_completions:
            role = msg.get("role", "")
            content = msg.get("content", "")
            content_clean = content[:300].replace("\n", " ")  # 清理换行符避免 f-string 报错
            print(f"[{role}]: {content_clean}...")

    if hasattr(step, "model_response") and step.model_response:
        model_resp_clean = step.model_response[:300].replace("\n", " ")
        print("\n🤖 Model Response:", model_resp_clean, "...")
```
